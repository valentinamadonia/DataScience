14) Â¿Cual es la expectativa de vida mas alta y mas baja? Usa la funcion de pandas y la funcion de NumPy y compara el resultado. Â¿Que paises tienen esas expectativas de vida?

```python
# Maxima expectativa de vida usando numpy
df2 = df[['Country','life_expectancy']].dropna()
array = df2.to_numpy()
maxNumpy= np.max(array[:, 1])
print("La expectativa de vida mas alta es: ", maxNumpy)
# Busco en el arreglo la expectativa de vida mas alta (segunda columna), y me quedo con los valores de la primera columna
paises = array[np.where(array[:, 1] == maxNumpy)][:, 0]
print(f"Los paises que tienen la expectativa de vida maxima de  {maxNumpy} son {', '.join(paises)}")
```

```python
# Minima expectativa de vida usando numpy
array2 = df2.to_numpy()
minNumpy= np.min(array2[:, 1])
print("La expectativa de vida mas baja es: ", minNumpy)
# Busco en el arreglo la expectativa de vida mas alta (segunda columna), y me quedo con los valores de la primera columna
paises2 = array2[np.where(array2[:, 1] == minNumpy)][:, 0]
print(f"Los paises que tienen la expectativa de vida minima de  {minNumpy} son {', '.join(paises2)}")
```

```python
# Maxima expectativa de vida usando pandas
max = df['life_expectancy'].max()
print(f"La expectativa de vida mas alta es {max}")
df5=df[df['life_expectancy'] == max]
df5 = df5['Country'].unique()
print(f"Los paises que tienen la expectativa de vida maxima de {max} son {', '.join(df5)}")
```

```python
# Minima expectativa de vida usando pandas
min = df['life_expectancy'].min()
print(f"La expectativa de vida mas baja es {min}")
df6=df[df['life_expectancy'] == min]
df6 = df6['Country'].unique()
print(f"Los paises que tienen la expectativa de vida minima de {min} son {', '.join(df6)}")
```
